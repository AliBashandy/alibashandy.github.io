```python
## Import Libraries

from bs4 import BeautifulSoup
import requests
import smtplib
import time
import datetime
```

```python
## Connect yo website

URL = 'https://shop.bodybuilding.com/products/bodybuilding-signature-whey-protein'
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36",
           #"Accept-Language": "en-US,en;q=0.9",
           "Accept-Encoding": "gzip, defelate, br", 
           "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
           "DNT":"1", 
           "Connection":"close", 
           "Upgrade-Insecure-Requests":"1"}

page = requests.get(URL, headers = headers)

soup1 = BeautifulSoup(page.content, "html.parser")

soup2 = BeautifulSoup(soup1.prettify(), "html.parser")

```

```python
print(soup2)
```

```python
## Get the title of the product
title = soup2.find("title").get_text(strip = True)
print(title)

## We have two prices one for memebers and one for non-members
non_memeber_price = soup2.find('div', class_ = 'product__price').get_text(strip = True)
print(non_memeber_price)

member_price = soup2.find(class_ = 'bf-price').get_text(strip = True)
print(member_price)

package_size_flavor = soup2.find(class_ = 'radio__legend__value').get_text()

## The size and flavor are detirmined by radio button we will get the chosen size and flavor
package_size_flavors = soup2.find_all(class_ = 'radio__legend__value')
for package_size_flavor in package_size_flavors:
    print(package_size_flavor.get_text(strip = True))
```
